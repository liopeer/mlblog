```python
import torch
import matplotlib.pyplot as plt

def p_y(z):
    return torch.sin(2*z) + 2*torch.sin(z)

z = torch.linspace(-6,6,100)
y = p_y(z)
noisy_y = y + 0.2*torch.randn_like(y)

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(z, y, label="ground truth label")
ax.plot(z, noisy_y, "o", label="noisy observed label")
ax.legend()
ax.set_xlabel("z")
ax.set_ylabel("y")
```
